```python
df = pd.read_csv('1,2,3월_1.csv')
```

```python
data = df[(df['업종명']=='카페')|(df['업종명']=='전시 및 행사 대행업')|(df['업종명']=='일반 음식점업')|(df['업종명']=='스포츠 및 레크레이션 용품 임대업')|(df['업종명']=='관광업')]
data
```

```python
result = data.groupby(['업종명'])['건수'].count().reset_index()
result
```
