```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("201901-202003.csv")
```

#### column 명 한글로 변경

```python
data.columns = ['년월','사용지_시도','사용지_시군구','업종명','거주지_시도','거주지_시군구','연령대','성별','생애주기','고객수','금액','건수']
```

#### 사용할 데이터만 뽑기
- 지역 : 대구, 경북
- 날짜 : 2019.1,2,3 and 2020.1,2,3

```python
data_dg = data[(data['사용지_시도'] == "대구")]
data_gb = data[(data['사용지_시도'] == "경북")]
```

```python
data_1 = data_dg.append(data_gb)
```

```python
data_1["년월"].value_counts()
```

```python
data_m = data_1[(data_1["년월"] == 201901) | (data_1["년월"] == 201902)|(data_1["년월"] == 201903)|(data_1["년월"] == 202001)
               |(data_1["년월"] == 202002)|(data_1["년월"] == 202003)]
data_m
```

#### 사용하지 않는 열 제외

```python
data_m1 = data_m.drop(["사용지_시도","사용지_시군구","거주지_시도","거주지_시군구"], axis = 1, inplace = False)
data_m1
```

#### 업종명 확인

```python
result = data_m1.groupby(['업종명'])['건수'].count().reset_index()
result
```

#### 7가지 업종 임의로 선택 : 일반 음식점업, 편의점, 관광업, 대중교통, 카페, 전시 및 행사 대행업, 스포츠 및 레크레이션 용품 임대업
- 특히
- 일식, 한식, 중식 : 일반음식점업
- 슈퍼마켓, 체인화 편의점 : 편의점
- 호텔업, 여행사업, 관광업, 정기 항공 운송업 : 관광업
- 버스 운송업 : 대중교통
- 비알콜 음료점업 : 카페
- 로 변경

```python
data_s1 = data_m1.replace({"일식 음식점업":"일반 음식점업","한식 음식점업":"일반 음식점업","중식 음식점업":"일반 음식점업","일반유흥 주점업":"일반 음식점업",
                            "슈퍼마켓":"편의점","체인화 편의점":"편의점",
                            "호텔업":"관광업","여행사업":"관광업","정기 항공 운송업":"관광업",
                           "버스 운송업":"대중교통", "비알콜 음료점업":"카페"})
```

```python
data_123 = data_s1[((data_s1['업종명'] == '일반 음식점업') | (data_s1['업종명'] == '편의점')) | (data_s1['업종명'] == "전시 및 행사 대행업")
                  |(data_s1["업종명"] == '관광업') | (data_s1["업종명"] == "카페")| (data_s1["업종명"] == "스포츠 및 레크레이션 용품 임대업")| (data_s1["업종명"] == "대중교통")]
```

```python
data_123.groupby(['년월',"업종명"])["건수"].count().reset_index()
```

```python
data_123.to_csv('1,2,3월_1.csv', encoding="euc-kr")
```

```python
data_m2 = data_123[(data_123["년월"] == 201902)|(data_123["년월"] == 201903)
               |(data_123["년월"] == 202002)|(data_123["년월"] == 202003)]
data_m2.groupby(["년월",'업종명'])["건수"].count().reset_index()
```
